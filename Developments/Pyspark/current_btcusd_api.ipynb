##Libraries

```python
import requests,json
from datetime import datetime
from pyspark.sql.functions import *

from pyspark.sql.types import *
```

##Functions

```python
def candlesapi_request(api_list):
    url = "https://api.coinbase.com/v2/exchange-rates?currency=BTC"
    headers = {"accept": "application/json"}
    response = requests.get(url, headers=headers)

    json_data = json.loads (response.text)
    api_list.append(((json_data['data'])['rates'])['USD'])
    
def save_candlesapi(timestamp,api_list):

    df_btcusd=spark.createDataFrame(api_list, StringType()).toDF('btcusd_price')
    
    
    df_btcusd=df_btcusd.withColumn('btcusd_price',df_btcusd.btcusd_price.cast('double'))
    df_btcusd=df_btcusd.withColumn('timestamp',to_timestamp(lit(timestamp), 'yyyy-MM-dd HH:mm:ss'))
    display(df_btcusd)

def current_btcusd_api():
    
    timestamp=datetime.now()
    timestamp=timestamp.strftime("%Y-%m-%d %H:%M:%S")
    
    api_list=[]
    candlesapi_request(api_list)
    save_candlesapi(timestamp,api_list)
```

```python
current_btcusd_api()
```
