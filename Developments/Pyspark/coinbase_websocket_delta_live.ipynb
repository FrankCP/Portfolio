```python
import dlt
import logging
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
@dlt.table(
    name = 'websocket_data_raw_table',
    comment=f'websocket raw data',
    table_properties = {
        'quality' : 'bronze',
        "pipelines.autoOptimize.managed":"true"
    }
)
def websocket_raw():
    logging.info(f"Reading websocket raw data")
    
    raw_table_schema=StructType([
        StructField('avg_px', StringType(), True),
        StructField('channel', StringType(), True),
        StructField('client_id', StringType(), True),
        StructField('client_order_id', StringType(), True),
        StructField('cum_qty', StringType(), True),
        StructField('fees', StringType(), True),
        StructField('leaves_qty', StringType(), True),
        StructField('order_id', StringType(), True),
        StructField('sequence_num', LongType(), True),
        StructField('status', StringType(), True),
        StructField('timestamp', StringType(), True),
        StructField('type', StringType(), True)
    ])
    
    df = spark \
        .readStream \
        .format("json") \
        .schema(raw_table_schema) \
        .option("recursiveFileLookup","true") \
        .option('ignoreChanges', 'true') \
        .load('Put here your dbfs path')
    return df
```
